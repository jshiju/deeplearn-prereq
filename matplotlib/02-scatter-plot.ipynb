```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(100,2) # 100 data points with 2 dimension
plt.scatter(X[:,0], X[:,1]);
```

```python
X = np.random.randn(200,2)
# By default the randn returns data points from standard normal centerd at zero
# skew half of the data by adding 3 so that it will be cenntered off os zero
X[:50] += 3
Y = np.zeros(200)
Y[:50] = 2
Y[50:100] = 3
plt.scatter(X[:,0], X[:,1], c=Y)

# c=Y - for color; taken from the one-dimensional array containing integers for different colors
```
